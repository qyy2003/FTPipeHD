```python
import pickle
import torch
import orjson
import os
```

```python
n=[200,300,20,20]
```

```python
a=torch.rand(n)
print(a)
```

```python
b=pickle.dumps(a)
```

```python
with open("pickle.txt", 'wb') as f:
    pickle.dump(a,f)
```

```python
# print(b)
print(len(b))
```

```python
a=a.tolist()
print(a)
```

```python
c=orjson.dumps(a)
```

```python
with open("orjson.orjson", 'w') as ff:
    orjson.dump(a,ff)
```

```python
# print(c)
print(len(c))
```
